```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm, linregress, t, expon, binom
import numpy as np
```

1:

```python
def generate():
    x = np.linspace(0, 1, 101)
    n=x.size
    f_x = 1 + 2 * x
    eps = norm.rvs(size = 101)
    y = f_x + eps

    reg = linregress(x, y)
    y_pred = reg.intercept + reg.slope * x

    std_samp = np.sqrt(((y - y_pred) ** 2).sum() / (n-2))
    x_std = x.std(ddof=0) * np.sqrt(n)
    return (reg.slope - 2) * x_std / std_samp
```

2:

```python
fs = pd.Series([generate() for i in range(5000)])
```

```python
df = 101 - 2
fig, ax = plt.subplots(1, 1)


x = np.linspace(t.ppf(0.01, df),
                t.ppf(0.99, df), 100)
fs.plot.density()
ax.plot(x, t.pdf(x, df),
       'r-', lw=5, alpha=0.6, label='t pdf')
plt.show()
```

```python
t_crit = t.ppf(1 - 0.05/2, df)
(np.abs(fs) >= t_crit).mean()
```
